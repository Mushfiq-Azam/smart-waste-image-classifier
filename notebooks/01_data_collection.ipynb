```python
from ddgs import DDGS
from fastai.vision.utils import download_images
import time

def scrape_images(search_terms, target_folder, max_per_term=25):
    target_folder.mkdir(parents=True, exist_ok=True)

    with DDGS() as ddgs:
        for term in search_terms:
            print(f"Searching: {term}")

            results = ddgs.images(
                query=term,
                max_results=max_per_term,
                safesearch="Moderate"
            )

            urls = [r["image"] for r in results]

            # ✅ FIX IS HERE
            download_images(dest=target_folder, urls=urls)

            time.sleep(10)  # avoid rate limiting
```

```python
plastic_terms = [
    "plastic bottle waste",
    "used plastic bottle trash",
    "plastic bottle recycling"
]

plastic_folder = RAW_DIR / "plastic_bottle"
scrape_images(plastic_terms, plastic_folder, max_per_term=25)
```

```python
from fastai.vision.utils import verify_images
from pathlib import Path

failed = verify_images(plastic_folder)
failed.map(Path.unlink)

print("Final image count:", len(list(plastic_folder.glob("*"))))
```

```python
import random
import matplotlib.pyplot as plt
from PIL import Image

# Only keep folders that are actual dataset classes
folders = sorted([
    f for f in RAW_DIR.iterdir()
    if f.is_dir() and not f.name.startswith(("test", "colab"))
])

# Limit to 12 classes
folders = folders[:12]

plt.figure(figsize=(15, 12))

for i, folder in enumerate(folders):
    images = list(folder.glob("*"))
    if len(images) == 0:
        continue

    img_path = random.choice(images)
    img = Image.open(img_path)

    plt.subplot(4, 3, i + 1)
    plt.imshow(img)
    plt.title(folder.name)
    plt.axis("off")

plt.tight_layout()
plt.show()
```
