```python
import random
import matplotlib.pyplot as plt
from PIL import Image

# Only dataset class folders
folders = sorted([
    f for f in RAW_DIR.iterdir()
    if f.is_dir()
])

# Limit to 12 classes
folders = folders[:12]

plt.figure(figsize=(15, 12))

for i, folder in enumerate(folders):
    images = list(folder.glob("*"))
    if len(images) == 0:
        continue

    img_path = random.choice(images)
    img = Image.open(img_path)

    plt.subplot(4, 3, i + 1)
    plt.imshow(img)
    plt.title(folder.name)
    plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
import random
import shutil

TRAIN_DIR = BASE_DIR / "data/train"
VALID_DIR = BASE_DIR / "data/valid"

TRAIN_DIR.mkdir(parents=True, exist_ok=True)
VALID_DIR.mkdir(parents=True, exist_ok=True)

random.seed(42)

for class_folder in RAW_DIR.iterdir():
    if not class_folder.is_dir():
        continue

    class_name = class_folder.name
    images = list(class_folder.glob("*"))
    random.shuffle(images)

    split_idx = int(0.8 * len(images))
    train_images = images[:split_idx]
    valid_images = images[split_idx:]

    (TRAIN_DIR / class_name).mkdir(parents=True, exist_ok=True)
    (VALID_DIR / class_name).mkdir(parents=True, exist_ok=True)

    for img in train_images:
        shutil.move(str(img), TRAIN_DIR / class_name / img.name)

    for img in valid_images:
        shutil.move(str(img), VALID_DIR / class_name / img.name)

    print(f"{class_name}: {len(train_images)} train | {len(valid_images)} valid")
```

```python
for class_dir in TRAIN_DIR.iterdir():
    train_count = len(list(class_dir.glob("*")))
    valid_count = len(list((VALID_DIR / class_dir.name).glob("*")))
    print(f"{class_dir.name}: {train_count} train | {valid_count} valid")
```

```python
from fastai.vision.utils import verify_images
from pathlib import Path

for split in ["train", "valid"]:
    split_dir = BASE_DIR / f"data/{split}"
    print(f"\nChecking {split.upper()} images")

    for folder in split_dir.iterdir():
        if folder.is_dir():
            failed = verify_images(folder)
            failed.map(Path.unlink)
            print(folder.name, "removed:", len(failed))
```

```python
from PIL import Image

def hard_image_check(folder):
    removed = 0
    for img_path in folder.glob("*"):
        try:
            img = Image.open(img_path)
            img.verify()
        except Exception:
            img_path.unlink()
            removed += 1
    return removed

for split in ["train", "valid"]:
    print(f"\nHard check: {split.upper()}")
    for folder in (BASE_DIR / f"data/{split}").iterdir():
        if folder.is_dir():
            r = hard_image_check(folder)
            print(folder.name, "removed:", r)
```
